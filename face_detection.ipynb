## Nesne Tespitinde Kullanılan cascade.xml dosyası
- Haarcascades dosyaları, istenilen nesneyi algılama için kullanılan, opencv nin hazır eğitilmiş dosyalarıdır.
- https://github.com/opencv/opencv/tree/4.x/data/haarcascades

```python
import matplotlib.pyplot as plt
import cv2
```

```python
einstein = cv2.imread('einstein.jpg',0)
plt.figure(), plt.imshow(einstein, cmap='gray'), plt.axis('off')
```

```python
# Yüz tanımlama için cascade classifier'ı yükle
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Yüzleri tespit et
face_rect = face_cascade.detectMultiScale(einstein)

# Rectangle'ı çiz
for (x, y, w, h) in face_rect:
    cv2.rectangle(einstein, (x, y), (x+w, y+h), (255, 0, 0), 10)

plt.figure(), plt.imshow(einstein, cmap='gray'), plt.axis('off')
```

## Multi-Face Recognition

```python
scientists = cv2.imread('scientists.jpg',0)
plt.figure(), plt.imshow(scientists, cmap='gray'), plt.axis('off')
```

```python
face_rect = face_cascade.detectMultiScale(scientists, minNeighbors=5) # minNeighborsu arttırarak false positive'ları azaltabiliriz

for (x, y, w, h) in face_rect:
    cv2.rectangle(scientists, (x, y), (x+w, y+h), (255, 0, 0), 3)
plt.figure(), plt.imshow(scientists, cmap='gray'), plt.axis('off')
```
