```python
import cv2 
import os 
import matplotlib.pyplot as plt
```

```python
# Kedileriin bulunduğu klasör
files = os.listdir('cats')
print(files)
```

```python
# Kedilerin konumları ile birlikte listeye ekliyoruz
img_path_list = []
for cat in files:
    img_path_list.append('cats/' + cat)
print(img_path_list)
```

```python
for cat in img_path_list:
    image = cv2.imread(cat)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Kedileri bulmak için cascade dosyasını yüklüyoruz
    cat_cascade = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
    rects = cat_cascade.detectMultiScale(image, scaleFactor=1.045,  minNeighbors=6)

    # Bulunan kedileri dikdörtgen içine alıyoruz
    for (i, (x, y, w, h)) in enumerate(rects):
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
        # Kedilerin üzerine numara yazıyoruz
        cv2.putText(image, "Cat #{}".format(i + 1), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.85, (0, 255, 255),2)

    plt.imshow(image),plt.axis('off')
    plt.show()
    """    
    CV2 ile görselleştirme
    cv2.imshow(cat, image)
    if cv2.waitKey(0) & 0xFF == ord('q'):
        continue
    """
```
